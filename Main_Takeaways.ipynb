#Main Takeaways
There is a lot of information condensed into this course but we hope that the main points: data order, memory/CPU speed, query optimization and data ingestion patterns, will be useful to you when faced with a real-time analytics project. 

Understanding what's really going on deep in the database is important. You should use the right tools for the job. Don't use PostgreSQL or MongoDB if you have a billion records. You arrive at the right answer by testing, understanding what is really happening and reading around the subject; this is what we do.

Please remember, don't forget:
 - Ingest as much data as you can at once 
 - Data ordering matters
 - Optimise queries and joins
 - Calculate on ingestion
