For this course, just keep in mind that Tinybird uses 
- Workspaces
- Data Sources - tables and views
- Pipes - a series of transformation nodes, where you can work with your data using SQL
- API Endpoints - to expose the result of a node of a pipe

These notebooks use the Tinybird CLI. Make sure you also explore your Workspace from the Tinybird UI to see the Dashboard and Data flow. In the Dashboard you can create data sources, design pipes and generate API endpoints interactively.
